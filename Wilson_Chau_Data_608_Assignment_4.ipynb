```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
url_provided = "https://raw.githubusercontent.com/Wilchau/Data_608_Assignment_4/main/DataAnalyst.csv"
df_provided = pd.read_csv(url_provided)

# Display the first few rows of the dataset
print(df_provided.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Plotting the distribution
plt.figure(figsize=(10, 6))
sns.histplot(df_provided['Salary'], kde=True, bins=30)
plt.title("Distribution of Salaries")
plt.xlabel("Salary")
plt.ylabel("Number of Jobs")
plt.show()
```

```python
# 2. Location Distribution for Data Analysts
avg_sal_by_location = df_provided.groupby('Location')['Salary'].mean().sort_values(ascending=False)

plt.figure(figsize=(15, 10))
avg_sal_by_location.plot(kind='bar', color='skyblue')
plt.title("Average Salary by Location")
plt.ylabel("Average Salary")
plt.xlabel("Location")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_sal_by_title = df_provided.groupby('Job Title')['Salary'].mean().sort_values(ascending=False).head(20)

plt.figure(figsize=(15, 10))
avg_sal_by_title.plot(kind='bar', color='lightgreen')
plt.title("Average Salary by Job Title (Top 20)")
plt.ylabel("Average Salary")
plt.xlabel("Job Title")
plt.xticks(rotation=45)
plt.show()
```
